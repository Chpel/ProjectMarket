# Окружение Детерминированное

```python
K_GROUPS = 1
K_CONS_GROUPS = 1
K_ADD_GROUPS = 1

func = ImpulseFunc(base=0., coef=1.7, level=0.3)
cons_list = [ListConsumer(K_GROUPS, K_CONS_GROUPS, list_budget=1.),
             ImpulseConsumer(K_GROUPS, K_ADD_GROUPS, func, add_budget=0.5)]
env = MarketEnv(K_GROUPS, fix_stock=10, consumer_list=cons_list, max_ep=3, action_type='discount')

fig, ax = plt.subplots(1,3, figsize=(15,6), width_ratios=[1,2,2])
policy_plot(env, ax[0]);
strategy_plot(env, 10, ax1=ax[1], ax2=ax[2]);
```
